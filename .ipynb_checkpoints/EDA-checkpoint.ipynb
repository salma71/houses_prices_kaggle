```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from scipy  import stats
from scipy.stats import norm, skew
```

```python
train = pd.read_csv('datasets/train.csv')
test = pd.read_csv('datasets/test.csv')
```

```python
train.head(5)
```

```python
train['SalePrice'].describe()
```

```python
sns.distplot(train['SalePrice'])
```

```python
# identify target variable -> SalePrice
def ecdf(data):
    # number of data  points
    n = len(data)
    # x - data
    x = np.sort(data)
    # y - data
    y = np.arange(1, n+1) / n
    
    return x, y
```

```python
x_price, y_price = ecdf(train['SalePrice'])

_ = plt.plot(x_price, y_price, marker='.', linestyle='none')

_ = plt.xlabel('Houses sale prices')
_ = plt.ylabel('ECDF')
```

```python
corrmat = train.corr()
plt.subplots(figsize=(12,9))
sns.heatmap(corrmat, vmax=0.9, square=True)
```

```python
#saleprice correlation matrix
k = 10 #number of variables for heatmap
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(train[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
#scatterplot
sns.set()
cols = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']
sns.pairplot(train[cols], size = 2.5)
plt.show()
```

```python
#missing data
total = train.isnull().sum().sort_values(ascending=False)
percent = (train.isnull().sum()/train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
# identify target variable -> SalePrice

# Draw ecdf for numerical variable, LotFrontage, LotArea, 
# yearBuilt,YearRemodAdd, MasVnrArea, BsmtFinSF1, BsmtFinSF2
# BsmtUnfSF, TotalBsmtSF, 1stFlrSF, 2ndFlrSF, LowQualFinSF,
# GrLivArea, GarageArea, 
def ecdf(data):
    # number of data  points
    n = len(data)
    # x - data
    x = np.sort(data)
    # y - data
    y = np.arange(1, n+1) / n
    
    return x, y
```

```python
# plot ECDF
x_fro, y_fro = ecdf(df['LotFrontage'])
x_area, y_area = ecdf(df['LotArea'])
_ = plt.plot(x_fro, y_fro, marker='.', linestyle='none')
_ = plt.plot(x_area, y_area, marker='.', linestyle='none')

_ = plt.xlabel('Linear feet of street connected to property LotFrontage')
_ = plt.ylabel('ECDF')
_ = plt.legend(('Lot'))
```
